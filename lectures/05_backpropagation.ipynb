In this lecture, the backprop algorithm will be explained in four different approaches. I personally learned the second approach first and it was stuck to me ever since then. The first approach introduced many unnecessary mathematical notations to explain the concept; I don't like it.

## 1. Pure Math
Suppose we have the following neural network. 

### Input Layer
$$
x = a^{(1)}
$$

### 2nd Layer (First Hidden)
$$
z^{(2)} = xW^{(1)} + b^{(1)}
a^{(2)} = \sigma(z^{(2)})
$$

### 3rd  Layer (Second Hidden)
$$
z^{(3)} = xW^{(2)} + b^{(2)}
a^{(3)} = \sigma(z^{(3)})
$$

### Output Layer
`U` is just another set of weight. The last layer does not have any nonlinearity.
$$
s = a^{(3)}U
$$
