## What is NLP?

NLP is a field at the intersection of computer science, artificial intelligence, and linguistics.

The goal is for computers to process or understand natural language in order to perform tasks that are useful, e.g.

* Making appointments
* Buying things online
* Question answering

Fully understanding and representing the meaning of language is a difficult gola. The perfect language understanding is AI-complete.

## What is Deep Learning?

Deep learning is a subfield of machine learning. Most machine learning methods work well because of human-designed representations and input features. 

For example, features for finding named entities like locations or organization names (Finkel et al., 2010)

**Representation learning** attempts to automatically learn good features or representations.

**Deep learning** algorithms attempt to learn multiple levels of representation and an output.
